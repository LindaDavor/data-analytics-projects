# TWO-WAY ANOVA

## Introduction

In the realm of statistical analysis, the two-way Analysis of Variance (ANOVA) stands as a robust methodological tool for examining the influence and interaction of two independent categorical variables on a continuous outcome variable. This inferential statistical test is pivotal in discerning not only the main effects of each independent variable but also the extent to which they collectively affect the dependent variable.

Employed extensively across various scientific disciplines, two-way ANOVA facilitates a comprehensive understanding of complex phenomena by allowing researchers to control variability and isolate the factorial contributions. The test’s capacity to evaluate the interaction effect is particularly invaluable, as it reveals whether the simultaneous manipulation of both factors leads to outcomes that are not predictable by considering each factor in isolation.

In essence, a two-way ANOVA serves as an indispensable instrument in the researcher’s toolkit, providing critical insights into the multifaceted dynamics at play within experimental and observational studies.


## Definition
Two-way analysis of variance (ANOVA) is a statistical method used to analyze the influence of two categorical independent variables (factors) on a continuous dependent variable. It allows us to determine whether there are significant differences in the means of the dependent variable across different levels of each factor, as well as whether there is an interaction effect between the two factors.

## Assumptions
**Independence:** Observations within each group are independent of each other. 
For example, if you’re testing the effectiveness of two teaching methods on student performance, each student’s score should be independent of others. If one student’s performance somehow affects another’s (maybe they study together), then the data points are not independent, and the results of the ANOVA could be misleading. To maintain independence, each student’s score should be based solely on their own abilities and the teaching method, not on other students’ performances.  

**Normality:** The dependent variable follows a normal distribution within each group.  
For instance, if you measure the heights of a large group of adults, you’ll find that most people are of average height, with fewer people being extremely tall or extremely short. When you plot these heights on a graph, they tend to form a bell-shaped curve, clustering around the average height with the frequencies tapering off towards the extremes.  

**Homogeneity of Variance:** The variance of the dependent variable is approximately equal across all groups. 
For example, imagine you’re comparing the heights of plants grown with different fertilizers. If one group of plants has heights that vary widely (some very tall, some very short) and another group has plants that are more similar in height, it would be unfair to compare the average heights directly. The group with more variation could have an average that’s less reliable.  

**No Significant Outliers:** There are no significant outliers in the data.  

## Examples

1. **Agricultural Research:** A researcher might want to understand the effect of different fertilizer types and planting densities on crop yield. By applying a two-way ANOVA, they can determine if there’s a significant difference in yield based on the combination of these two factors.  
 
2. **Educational Studies:** Educators could use a two-way ANOVA to explore the interaction between gender and educational level on test anxiety among university students. This helps to see if the level of anxiety varies not just with gender or education level alone, but with the specific combinations of both.  

3. **Botanical Experiments:** A botanist may use a two-way ANOVA to assess how sunlight exposure and watering frequency affect plant growth. This analysis can reveal if these factors independently influence growth, or if there’s a combined effect when varying both sunlight and water together.  

4. **Product Testing:** Companies might conduct a two-way ANOVA to evaluate how two different variables, such as temperature settings and material types, affect the durability of a product. This can inform if certain materials are more resistant to temperature changes than others.

5. **Marketing Research:** A company might want to test the effectiveness of different advertising campaigns on two different age groups. They could use a two-way ANOVA to analyze the impact of campaign type and age group on consumer purchasing behavior

```python
# Sample Code
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt

# Sample data: Plant growth dataset
data = pd.DataFrame({
    'Fertilizer': np.repeat(['Type1', 'Type2'], 15),
    'Watering': np.tile(np.repeat(['Low', 'Medium', 'High'], 5), 2),
    'Height': [6, 6, 7, 5, 6, 5, 5, 6, 4, 5, 6, 6, 7, 8, 7,
               3, 4, 4, 4, 5, 4, 4, 4, 4, 4, 5, 6, 6, 7, 8]
})

# Perform two-way ANOVA
model = ols('Height ~ C(Fertilizer) + C(Watering) + C(Fertilizer):C(Watering)', data=data).fit()
anova_results = sm.stats.anova_lm(model, typ=2)

# Print ANOVA table
print(anova_results)

# Interaction plot for the two-way ANOVA
interaction_plot = sm.graphics.interaction_plot(data['Watering'], data['Fertilizer'], data['Height'], 
                                                colors=['red', 'blue'], markers=['D', '^'], ms=10)

plt.show()
```

Here’s a simple breakdown of the code and its components:

**Importing Libraries:** The code begins by importing necessary Python libraries that contain functions and methods to perform statistical analysis and data visualization.  

**Creating Data:** A dataset is created using pandas, which simulates the growth of plants based on different types of fertilizers and watering levels. The ‘Height’ column represents the growth outcome.  

**ANOVA Model:** The ols function from statsmodels is used to create a model that describes the relationship between the plant heights and the two factors: type of fertilizer and watering level.  

**ANOVA Results:** The anova_lm function then performs the ANOVA test, which results in a table showing whether the differences in plant heights are statistically significant based on the different treatments.  

**Adding Visualization:** The added code for visualization uses matplotlib to create an interaction plot. This plot will show how plant height varies with each combination of fertilizer type and watering level.  

**Displaying the Plot:** Finally, the plt.show() command displays the plot, which helps in visually understanding the interaction between the two factors.  

The resulting visualization will help you see if there’s a pattern or trend in plant growth based on the combination of fertilizer and watering. For example, you might see that ‘Type1’ fertilizer with ‘High’ watering results in taller plants, which would be indicated by higher points on the plot for that group. The different colors and markers make it easy to distinguish between the groups in the plot. This visual aid can be very helpful in quickly identifying which combinations of treatments are most effective for plant growth.
